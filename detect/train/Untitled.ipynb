```python
from ultralytics import YOLO
import cv2

# Load model
model = YOLO(r"C:\Users\shah\OneDrive\Desktop\ML projects\best.pt")

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Predict
    results = model(frame)

    # Plot results
    annotated_frame = results[0].plot()

    # Show window
    cv2.imshow("YOLO Real-Time", annotated_frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
